```python
import pandas as pd

df = pd.read_csv("Nassau Candy Distributor.csv")
df.head()
```

```python
df[df['Sales'] <= 0]
```

```python
df[(df['Units'].isnull()) | (df['Units'] <= 0)]
```

```python
df[df['Cost'] > df['Sales']]
```

```python
df_clean = df[(df['Sales'] > 0) & (df['Units'] > 0)]
```

Cost greater than sales records were retained to identify margin risk products.

```python
df_clean['Division'] = df_clean['Division'].str.strip().str.title()
df_clean['Division'].value_counts()
```

```python
df_clean.isnull().sum()
df_clean.describe()
```

## Data Cleaning Summary

The dataset was cleaned to ensure accurate profitability analysis.

Actions taken:
- Removed records with zero or negative sales
- Removed records with missing or zero unit values
- Standardized division names for consistency
- Retained loss-making products (Cost > Sales) for margin risk analysis

This ensures that all profitability metrics are financially meaningful.
